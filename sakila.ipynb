# Sakila database analysis

This Jupyter notebook analyses the MySQL Sakila database 

```python
### This function asks for a database name to work whith and assign the value to a variable
def askdatabase():
    database = input("What MySQL database you want to work whith? ")
    return database
```

```python
global database_name
database_name = askdatabase()
```

```python
def queryexecutor(args):
    """This function executes as many queries as in the LIST passed as arguments from the invoking statement."""
    from getpass import getpass
    from mysql.connector import connect, Error
    
    try:
        with connect(
            host = "localhost",
            user = input("Enter username: "),
            password = getpass("Enter password: "),
            database = database_name #input("Tell me the database do you want to use: "),
        ) as connection:
                with connection.cursor() as cursor:
                    for arg in args:
                        print("The current query is :", arg, "\n")
                        cursor.execute(arg)
                        result = cursor.fetchall()
                        for row in result:
                            print(row)
    except Error as e:
        print(e)
    return result
```

```python
query = ("""select f.film_id, f.title, f.release_year, f.length, f.rating, c.name, l.name
from film as f
join film_category as fc on fc.film_id = f.film_id
join category as c on c.category_id = fc.category_id
join language as l on l.language_id = f.language_id;""")
result = queryexecutor([query])
```

```python
import pandas as pd
df = pd.DataFrame(result, columns = ['film_id','title', 'release_year', 'length','rating', 'category','language'])
```
